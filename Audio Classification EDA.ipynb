```python
filename = 'UrbanSound8K/dog_bark.wav'
```

```python
import IPython.display as ipd
```

```python
import librosa 
import librosa.display
```

```python
#checking an example of dog sound
plt.figure(figsize=(14, 5))
#use librosa library since the sample rate of the raw data should be an integer
#get data and sample rate of the sound data
data, sample_rate = librosa.load(filename)
librosa.display.waveshow(data, sr = sample_rate)
ipd.Audio(filename='UrbanSound8K/dog_bark.wav')
```

In audio production, the sample rate defines the number of times per second a sound is sampled. Generally, the standard sample rate used for audio CDs is around 44 kH

```python
#let's see sample rate and data of the dog bark
sample_rate # the sample rate here is around 22 kH
```

When reading this info with librosa, we are reading the signals with the sample rate of 22 kH. We can even use skipy library with audio data. 

Audio has 2 channels: stereo and mono. In mono, the signals will be reaching equally to both ears. In stereo, different audio signals may reach both ears differently at different times.

Librosa will normalize the data and output it with a single sample rate.

```python
from scipy.io import wavfile as wav
wave_sample_rate, wave_audio=wav.read(filename)
```

```python
wave_sample_rate #here the sample rate is 44.1 kH, which is the standard sample rate for audio CDs.
#Librosa is bringing a mechanism to read the audio file and convert it to a sample rate of 22 kH
```

```python
wave_audio #these array values indicate the integer values of each sound wave. Combining these values will create a sound wave.
```
